# Chapter 9: Coarseness of Coarse Coding

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')

x = np.linspace(0, 1, 400)
coarse = ((x > 0.2) & (x < 0.8)).astype(float)
fine = ((x > 0.45) & (x < 0.55)).astype(float)
plt.plot(x, coarse, label='coarse receptive field')
plt.plot(x, fine, label='fine receptive field')
plt.legend()
plt.title('Coarseness in coarse coding')
plt.show()
```
