# Sutton & Barto Chapter 3: Policy Improvement by One-Step Lookahead
## Turning Values Into Better Decisions

This notebook focuses on a small but important idea: if we already have useful state values, we can look one step ahead and choose better actions.

## What You Will See

- one-step lookahead action scoring
- greedy improvement from a value table
- a comparison between a soft policy and its greedy improvement

Dependencies: `numpy`

## Theory

Given a state-value estimate $v(s)$, we can score each action by:

$$
\sum_{s', r} p(s', r \mid s, a)[r + \gamma v(s')].
$$

A greedy improvement step picks the action with the highest score. This does not require solving the whole problem again. It only needs one-step dynamics and the current value estimate.

This basic idea later becomes the backbone of policy iteration and many model-based methods.

```python
gamma = 0.9

value_estimate = {'A': 1.0, 'B': 2.0, 'T': 0.0}

transition_model = {
    ('A', 'left'): [('A', -1.0, 1.0)],
    ('A', 'right'): [('B', 0.0, 1.0)],
    ('B', 'left'): [('A', 0.0, 1.0)],
    ('B', 'right'): [('T', 2.0, 1.0)],
}

old_policy = {
    'A': {'left': 0.5, 'right': 0.5},
    'B': {'left': 0.5, 'right': 0.5},
}
```

```python
def one_step_lookahead(state):
    scores = {}
    for action in ['left', 'right']:
        total = 0.0
        for next_state, reward, prob in transition_model[(state, action)]:
            total += prob * (reward + gamma * value_estimate[next_state])
        scores[action] = total
    return scores

lookahead_scores = {s: one_step_lookahead(s) for s in ['A', 'B']}
lookahead_scores
```

```python
greedy_policy = {s: max(scores, key=scores.get) for s, scores in lookahead_scores.items()}
greedy_policy
```

## Takeaway

- One-step lookahead converts a state-value table into action preferences.
- Greedy improvement is local in computation but powerful in effect.
- This is one of the cleanest conceptual bridges between value estimation and control.
