This notebook focuses on the return definition in Chapter 3 and why discounting changes the behavior that a policy prefers.

- why delayed rewards are worth less when $\gamma < 1$

Dependencies: `numpy`, `matplotlib`

## Theory

The return from time $t$ is the total future reward from that point onward:

$$
G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \cdots
$$

When $\gamma = 1$, the agent values future rewards the same as immediate rewards in an episodic task.

When $\gamma < 1$, rewards far in the future have less influence. This is useful both mathematically and conceptually:

- long delays matter less
- infinite-horizon sums are better behaved
- near-term outcomes get more emphasis


```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
```

```python
rewards_fast = [0, 5]
rewards_slow = [-1, 0, 5]
gammas = [1.0, 0.9, 0.5]

def discounted_return(rewards, gamma):
    return sum((gamma ** t) * r for t, r in enumerate(rewards))

for gamma in gammas:
    print(f'gamma={gamma:.1f} | fast={discounted_return(rewards_fast, gamma):.3f} | slow={discounted_return(rewards_slow, gamma):.3f}')
```

```python
gamma_grid = np.linspace(0.1, 1.0, 50)
fast_values = [discounted_return(rewards_fast, g) for g in gamma_grid]
slow_values = [discounted_return(rewards_slow, g) for g in gamma_grid]

fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(gamma_grid, fast_values, label='fast path rewards [0, 5]')
ax.plot(gamma_grid, slow_values, label='slow path rewards [-1, 0, 5]')
ax.set_title('Effect of Discounting on Return')
ax.set_xlabel('gamma')
ax.set_ylabel('discounted return')
ax.legend()
plt.tight_layout()
```

## Takeaway

- Return compresses a whole reward sequence into one number.
- Discounting changes the ranking of behaviors when one path delays reward more than another.
- Chapter 3 uses this return definition as the target for value functions.
