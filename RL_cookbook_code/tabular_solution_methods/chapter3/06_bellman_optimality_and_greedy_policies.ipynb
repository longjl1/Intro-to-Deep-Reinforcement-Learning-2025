# Sutton & Barto Chapter 3: Bellman Optimality and Greedy Policies
## How Optimal Values Select Better Actions

This notebook moves from values under a fixed policy to the idea of **optimal value functions** and greedy improvement.

## What You Will See

- Bellman optimality equation for $v_*$
- Bellman optimality equation for $q_*$
- greedy action selection from action values
- a tiny worked example

Dependencies: `numpy`

## Theory

The optimal state-value function is:

$$
v_*(s) = \max_a \sum_{s', r} p(s', r \mid s, a)[r + \gamma v_*(s')].
$$

The optimal action-value function is:

$$
q_*(s, a) = \sum_{s', r} p(s', r \mid s, a) \left[r + \gamma \max_{a'} q_*(s', a')\right].
$$

These equations say that an optimal value function already assumes the best future choices will be made.

A **greedy policy** with respect to a value estimate chooses the action with the largest one-step lookahead value.

```python
import numpy as np
```

```python
states = ['A', 'B', 'T']
non_terminal = ['A', 'B']
actions = ['left', 'right']
gamma = 0.9

transitions = {
    ('A', 'left'): ('A', -1.0),
    ('A', 'right'): ('B', 0.0),
    ('B', 'left'): ('A', 0.0),
    ('B', 'right'): ('T', 2.0),
}
```

```python
def optimality_update(v):
    new_v = v.copy()
    for s in non_terminal:
        action_returns = []
        for a in actions:
            next_state, reward = transitions[(s, a)]
            action_returns.append(reward + gamma * v[next_state])
        new_v[s] = max(action_returns)
    return new_v


v_star = {s: 0.0 for s in states}
for _ in range(20):
    v_star = optimality_update(v_star)
v_star
```

```python
q_star = {}
greedy_policy = {}

for s in non_terminal:
    action_values = {}
    for a in actions:
        next_state, reward = transitions[(s, a)]
        action_values[a] = reward + gamma * v_star[next_state]
        q_star[(s, a)] = action_values[a]
    greedy_policy[s] = max(action_values, key=action_values.get)

q_star, greedy_policy
```

## Interpretation

In this tiny example, the greedy policy should learn to move toward the terminal reward rather than loop with negative reward. The Bellman optimality equation encodes that preference directly through the `max` operator.

## Takeaway

- Bellman expectation equations describe a fixed policy.
- Bellman optimality equations replace policy averaging with a `max` over actions.
- Greedy improvement is the bridge from value estimates to better decision rules.
