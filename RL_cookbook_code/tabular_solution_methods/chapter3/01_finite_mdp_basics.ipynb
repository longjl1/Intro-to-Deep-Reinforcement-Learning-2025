## Theory

A finite MDP is usually described by:

- a finite state set $\mathcal{S}$
- a finite action set $\mathcal{A}$
- transition dynamics $p(s', r \mid s, a)$
- a discount factor $\gamma \in [0, 1]$

The **Markov property** says that the next-state and reward distribution depends only on the current state and action, not on the whole past history.

The discounted return from time $t$ is:

$$
G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \cdots
$$

When $\gamma < 1$, future rewards are geometrically discounted.

```python
import numpy as np
```

```python
states = ['start', 'middle', 'goal']
actions = ['move', 'wait']
gamma = 0.9

# p(s', r | s, a) is represented here as a small dictionary of outcomes.
transitions = {
    ('start', 'move'): [('middle', 0.0, 1.0)],
    ('start', 'wait'): [('start', -1.0, 1.0)],
    ('middle', 'move'): [('goal', 5.0, 1.0)],
    ('middle', 'wait'): [('middle', -0.5, 1.0)],
}

for key, outcomes in transitions.items():
    print(f'{key} -> {outcomes}')
```

## A small return calculation

Suppose an episode produces rewards `[0, 5]` after leaving `start` and then reaching `goal`. The discounted return is easy to compute directly.

```python
rewards = [0.0, 5.0]
discounted_return = sum((gamma ** t) * r for t, r in enumerate(rewards))
discounted_return
```

If instead the agent wastes one extra step and gets rewards `[-1, 0, 5]`, the return becomes smaller because the positive reward arrives later and the early penalty is immediate.

```python
delayed_rewards = [-1.0, 0.0, 5.0]
delayed_return = sum((gamma ** t) * r for t, r in enumerate(delayed_rewards))
delayed_return
```

## Takeaway

- A finite MDP gives a precise probabilistic model of sequential decision making.
- The Markov property lets us reason from the current state instead of the full history.
- Discounted return turns a whole future reward sequence into one scalar objective.
