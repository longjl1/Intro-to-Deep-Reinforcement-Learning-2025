# Sutton & Barto Chapter 3: Bellman Expectation Equations
## State Values and Action Values Under a Policy

This notebook introduces the **Bellman expectation equation** for a fixed policy in a finite MDP.

## What You Will See

- a tiny deterministic finite MDP
- state-value function $v_\pi(s)$
- action-value function $q_\pi(s, a)$
- Bellman expectation updates by direct calculation

Dependencies: `numpy`

## Theory

For a policy $\pi$, the state-value function is the expected return starting from state $s$ and following $\pi$:

$$
v_\pi(s) = \mathbb{E}_\pi[G_t \mid S_t = s].
$$

The Bellman expectation equation rewrites this recursively:

$$
v_\pi(s) = \sum_a \pi(a \mid s) \sum_{s', r} p(s', r \mid s, a) [r + \gamma v_\pi(s')].
$$

Similarly, the action-value function is:

$$
q_\pi(s, a) = \sum_{s', r} p(s', r \mid s, a) [r + \gamma \sum_{a'} \pi(a' \mid s') q_\pi(s', a')].
$$

These equations are central because they connect values to one-step lookahead.

```python
import numpy as np
```

```python
states = ['A', 'B', 'T']
non_terminal = ['A', 'B']
actions = ['left', 'right']
gamma = 0.9

transitions = {
    ('A', 'left'): ('A', -1.0),
    ('A', 'right'): ('B', 0.0),
    ('B', 'left'): ('A', 0.0),
    ('B', 'right'): ('T', 2.0),
}

policy = {
    'A': {'left': 0.5, 'right': 0.5},
    'B': {'left': 0.5, 'right': 0.5},
}
```

```python
def bellman_expectation_update(v):
    new_v = v.copy()
    for s in non_terminal:
        total = 0.0
        for a, prob_a in policy[s].items():
            next_state, reward = transitions[(s, a)]
            total += prob_a * (reward + gamma * v[next_state])
        new_v[s] = total
    return new_v


v = {s: 0.0 for s in states}
for _ in range(20):
    v = bellman_expectation_update(v)
v
```

## From state values to action values

Once we have an estimate of $v_\pi$, it is easy to compute $q_\pi(s, a)$ from one-step lookahead.

```python
q = {}
for s in non_terminal:
    for a in actions:
        next_state, reward = transitions[(s, a)]
        q[(s, a)] = reward + gamma * v[next_state]

q
```

## Linear-system viewpoint

For a fixed policy, the Bellman expectation equation is linear in the unknown state values. In very small problems, we could solve that linear system directly. In practice, iterative updates are often more intuitive and connect naturally to later dynamic programming methods.

That is why the recursive form is so important: it is both a mathematical identity and a computational update rule.

## Takeaway

- The Bellman expectation equation turns long-horizon return into a recursive one-step update.
- State values and action values are tightly connected.
- This recursion is the foundation for later dynamic programming and reinforcement learning algorithms.
