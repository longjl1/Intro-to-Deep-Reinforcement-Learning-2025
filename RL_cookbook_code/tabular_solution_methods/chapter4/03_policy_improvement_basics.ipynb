# Chapter 4: Policy Improvement Basics

Once a policy has been evaluated, we can improve it greedily with respect to its state values.

## Policy Improvement Theorem

If a new policy chooses, in each state, an action that is at least as good as the old policy's expected action, then the new policy is no worse overall.

This is the main bridge from evaluation to control.

```python
gamma = 0.9
v = {'A': 1.0, 'B': 2.0, 'T': 0.0}
transitions = {
    ('A', 'left'): ('A', -1.0),
    ('A', 'right'): ('B', 0.0),
    ('B', 'left'): ('A', 0.0),
    ('B', 'right'): ('T', 2.0),
}

greedy = {}
for s in ['A', 'B']:
    scores = {}
    for a in ['left', 'right']:
        ns, r = transitions[(s, a)]
        scores[a] = r + gamma * v[ns]
    greedy[s] = max(scores, key=scores.get)
greedy
```
