# Chapter 4: Asynchronous DP and In-Place Updates

Dynamic programming does not require sweeping every state in lockstep. In-place and asynchronous updates can still converge and are often cheaper in practice.

```python
gamma = 0.9
v = {'A': 0.0, 'B': 0.0, 'T': 0.0}
transitions = {
    ('A', 'left'): ('A', -1.0),
    ('A', 'right'): ('B', 0.0),
    ('B', 'left'): ('A', 0.0),
    ('B', 'right'): ('T', 2.0),
}
order = [('A', 'right'), ('B', 'right'), ('A', 'right'), ('B', 'right')]
for s, a in order:
    ns, r = transitions[(s, a)]
    v[s] = r + gamma * v[ns]
v
```

## Why This Matters

In-place updates let later states in a sweep use fresher values immediately. That often speeds convergence and previews practical dynamic programming implementations.
