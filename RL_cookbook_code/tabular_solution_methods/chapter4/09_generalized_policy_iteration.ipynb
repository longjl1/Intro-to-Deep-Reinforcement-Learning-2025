# Chapter 4: Generalized Policy Iteration

Generalized policy iteration (GPI) is the idea that evaluation and improvement can proceed together instead of as strictly separated full phases.

## Intuition

- policy evaluation pushes values toward correctness
- policy improvement pushes actions toward greediness
- alternating or interleaving both tends to produce better policies


```python
gpi_loop = [
    'evaluate current policy a little',
    'improve policy a little',
    'repeat until both stabilize',
]
gpi_loop
```

Policy iteration and value iteration can both be viewed as specific instances of this broader GPI picture.
