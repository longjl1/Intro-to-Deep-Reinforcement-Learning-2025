# Chapter 4: Policy Iteration on a Tiny MDP

```python
gamma = 0.9
states = ['A', 'B', 'T']
actions = ['left', 'right']
transitions = {
    ('A', 'left'): ('A', -1.0),
    ('A', 'right'): ('B', 0.0),
    ('B', 'left'): ('A', 0.0),
    ('B', 'right'): ('T', 2.0),
}

policy = {'A': 'left', 'B': 'left'}
for _ in range(5):
    v = {s: 0.0 for s in states}
    for _ in range(20):
        new_v = v.copy()
        for s in ['A', 'B']:
            a = policy[s]
            ns, r = transitions[(s, a)]
            new_v[s] = r + gamma * v[ns]
        v = new_v
    stable = True
    for s in ['A', 'B']:
        old = policy[s]
        policy[s] = max(actions, key=lambda a: transitions[(s, a)][1] + gamma * v[transitions[(s, a)][0]])
        stable = stable and (old == policy[s])
    if stable:
        break
policy, v
```
