# Chapter 4: Value Iteration on a Tiny MDP

Value iteration combines a truncated evaluation step with immediate greedy improvement:

$$
v_{k+1}(s) = \max_a \sum_{s', r} p(s', r\mid s, a)[r + \gamma v_k(s')].
$$


```python
gamma = 0.9
states = ['A', 'B', 'T']
actions = ['left', 'right']
transitions = {
    ('A', 'left'): ('A', -1.0),
    ('A', 'right'): ('B', 0.0),
    ('B', 'left'): ('A', 0.0),
    ('B', 'right'): ('T', 2.0),
}

v = {s: 0.0 for s in states}
for _ in range(20):
    new_v = v.copy()
    for s in ['A', 'B']:
        new_v[s] = max(transitions[(s, a)][1] + gamma * v[transitions[(s, a)][0]] for a in actions)
    v = new_v
policy = {s: max(actions, key=lambda a: transitions[(s, a)][1] + gamma * v[transitions[(s, a)][0]]) for s in ['A', 'B']}
policy, v
```

## Interpretation

Value iteration keeps only the best one-step backup in each state. In this tiny example, that quickly drives the values toward the path that reaches the terminal reward.
