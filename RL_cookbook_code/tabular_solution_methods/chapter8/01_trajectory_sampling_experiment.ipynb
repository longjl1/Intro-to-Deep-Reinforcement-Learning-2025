# Chapter 8: Trajectory Sampling Experiment

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')

rng = np.random.default_rng(0)
steps = np.arange(1, 101)
uniform = np.log1p(steps) / np.log(101)
trajectory = 1 - np.exp(-steps / 25)

plt.plot(steps, uniform, label='uniform updates')
plt.plot(steps, trajectory, label='trajectory sampling')
plt.xlabel('planning updates')
plt.ylabel('relative performance')
plt.title('Illustrative trajectory sampling comparison')
plt.legend()
plt.show()
```

This notebook uses a lightweight synthetic comparison rather than reproducing the exact original experiment. The purpose is to illustrate why focused trajectory-based planning can outperform broader but shallower uniform updates.
