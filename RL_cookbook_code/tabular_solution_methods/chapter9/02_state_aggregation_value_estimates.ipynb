# Chapter 9: Visualizing State Aggregation Value Estimates

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')

x = np.arange(1000)
true_curve = x / 999
approx_curve = np.repeat(np.linspace(0.05, 0.95, 10), 100)

plt.plot(x, true_curve, label='smooth reference')
plt.step(x, approx_curve, where='mid', label='state aggregation')
plt.legend()
plt.title('Piecewise-constant state aggregation approximation')
plt.show()
```
