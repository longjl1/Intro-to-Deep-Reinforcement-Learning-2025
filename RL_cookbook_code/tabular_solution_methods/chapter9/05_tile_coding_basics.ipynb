# Chapter 9: Tile Coding Basics

```python
import numpy as np

def tile_indices(x, n_tilings=4, bins=8):
    idx = []
    for tiling in range(n_tilings):
        offset = tiling / (n_tilings * bins)
        idx.append(min(bins - 1, int((x + offset) * bins)))
    return idx

tile_indices(0.37)
```

Tile coding keeps generalization local while still allowing multiple overlapping active features. That makes it especially useful for control tasks like Mountain Car.
