# Chapter 9: State Aggregation on the 1000-state Random Walk

```python
import numpy as np

n_states = 1000
n_groups = 10
group_size = n_states // n_groups
weights = np.zeros(n_groups)

def feature_index(state):
    return min(n_groups - 1, state // group_size)

sample_states = [0, 99, 100, 599, 999]
[(s, feature_index(s)) for s in sample_states]
```
