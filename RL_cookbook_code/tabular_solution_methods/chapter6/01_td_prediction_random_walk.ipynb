# Chapter 6: TD Prediction on the Random Walk

A minimal Python implementation of TD(0) on the classic random walk example.

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')

states = np.arange(7)
true_values = np.array([0, 1/6, 2/6, 3/6, 4/6, 5/6, 0])

def generate_episode(rng):
    s = 3
    episode = []
    while s not in (0, 6):
        ns = s + rng.choice([-1, 1])
        reward = 1.0 if ns == 6 else 0.0
        episode.append((s, reward, ns))
        s = ns
    return episode

rng = np.random.default_rng(0)
v = np.full(7, 0.5)
v[0] = v[6] = 0.0
alpha = 0.1
history = []
for _ in range(100):
    ep = generate_episode(rng)
    for s, r, ns in ep:
        v[s] += alpha * (r + v[ns] - v[s])
    history.append(v.copy())

plt.plot(np.array(history)[:, 1:6])
plt.plot(true_values[1:6], 'ko--', label='true values')
plt.title('TD(0) estimates on the random walk')
plt.legend()
plt.show()
v
```

TD prediction updates after each transition instead of waiting for the full episode return. That is the key contrast with Monte Carlo prediction.
