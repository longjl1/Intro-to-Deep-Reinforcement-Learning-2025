# Chapter 6: Alternate TD Random Walk Implementation

```python
import numpy as np

class RandomWalk:
    def __init__(self, n=5, seed=0):
        self.n = n
        self.rng = np.random.default_rng(seed)
        self.start = (n + 1) // 2

    def episode(self):
        s = self.start
        out = []
        while s not in (0, self.n + 1):
            ns = s + self.rng.choice([-1, 1])
            reward = 1.0 if ns == self.n + 1 else 0.0
            out.append((s, reward, ns))
            s = ns
        return out

env = RandomWalk()
values = np.full(env.n + 2, 0.5)
values[0] = values[-1] = 0.0
for _ in range(50):
    for s, r, ns in env.episode():
        values[s] += 0.05 * (r + values[ns] - values[s])
values
```
