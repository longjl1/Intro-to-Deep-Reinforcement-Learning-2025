# Chapter 5: Epsilon-Soft On-Policy Monte Carlo Control

Instead of relying on exploring starts, we can maintain an epsilon-soft policy so every action continues to have some chance of being sampled.

A common form is:

$$
\pi(a\mid s)=
\begin{cases}
1-\epsilon+\frac{\epsilon}{|\mathcal{A}(s)|} & \text{if } a \text{ is greedy} \\
\frac{\epsilon}{|\mathcal{A}(s)|} & \text{otherwise}
\end{cases}
$$
