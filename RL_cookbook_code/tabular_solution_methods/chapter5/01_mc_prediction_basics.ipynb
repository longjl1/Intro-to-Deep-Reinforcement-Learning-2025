# Chapter 5: Monte Carlo Prediction Basics

Monte Carlo prediction estimates value functions from complete episodes, without needing the environment model.

```python
episodes = [
    [('A', 0), ('B', 2)],
    [('A', 1)],
    [('A', 0), ('B', 1)],
]

returns = {'A': [], 'B': []}
for ep in episodes:
    G = 0
    for state, reward in reversed(ep):
        G += reward
        returns[state].append(G)
values = {s: sum(gs) / len(gs) for s, gs in returns.items()}
values
```
