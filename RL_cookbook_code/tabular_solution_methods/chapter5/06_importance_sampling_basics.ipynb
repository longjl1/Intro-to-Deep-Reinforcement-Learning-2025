# Chapter 5: Importance Sampling Basics

Importance sampling reweights returns collected under one policy so they can estimate values for another policy.

```python
target_prob = 0.8
behavior_prob = 0.5
rho = target_prob / behavior_prob
rho
```

If a full trajectory is involved, the total importance ratio becomes a product of per-step ratios. That is what makes off-policy Monte Carlo flexible, but it is also why variance can grow quickly.

```python
per_step_ratios = [1.2, 0.8, 1.1]
trajectory_ratio = 1.0
for ratio in per_step_ratios:
    trajectory_ratio *= ratio
trajectory_ratio
```
