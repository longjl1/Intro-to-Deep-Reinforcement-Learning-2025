# Chapter 5: Ordinary vs Weighted Importance Sampling

```python
weighted_numerator = [1.6, 0.8, 2.4]
weights = [1.6, 0.4, 1.2]
ordinary = sum(weighted_numerator) / len(weighted_numerator)
weighted = sum(weighted_numerator) / sum(weights)
ordinary, weighted
```

Weighted importance sampling often trades a little bias for lower variance in finite samples. That is why it is frequently preferred in practice for Monte Carlo off-policy estimation.
