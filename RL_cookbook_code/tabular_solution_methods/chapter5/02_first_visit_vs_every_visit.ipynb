# Chapter 5: First-Visit vs Every-Visit Monte Carlo

```python
episode = [('A', 0), ('A', 1), ('B', 2)]

first_visit_returns = {}
every_visit_returns = {}
seen = set()
G = 0
for t in range(len(episode) - 1, -1, -1):
    state, reward = episode[t]
    G += reward
    every_visit_returns.setdefault(state, []).append(G)
    if state not in seen:
        first_visit_returns.setdefault(state, []).append(G)
        seen.add(state)

first_visit_returns, every_visit_returns
```

## Interpretation

First-visit Monte Carlo uses only the first occurrence of each state in an episode. Every-visit Monte Carlo uses all occurrences. Both are consistent under standard assumptions, but they weight samples differently.
