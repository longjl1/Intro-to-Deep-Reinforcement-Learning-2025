# Chapter 7: Comparing n-step TD Settings on the Random Walk

```python
import numpy as np

true_values = np.array([0, 1/6, 2/6, 3/6, 4/6, 5/6, 0])
estimates = {
    1: np.array([0, 0.18, 0.31, 0.48, 0.67, 0.81, 0]),
    2: np.array([0, 0.15, 0.29, 0.50, 0.69, 0.83, 0]),
    4: np.array([0, 0.12, 0.27, 0.49, 0.71, 0.85, 0]),
}
rmse = {n: np.sqrt(np.mean((v[1:6] - true_values[1:6])**2)) for n, v in estimates.items()}
rmse
```

Changing the step length changes the bias-variance tradeoff. Small n behaves more like TD, while large n moves closer to Monte Carlo.
