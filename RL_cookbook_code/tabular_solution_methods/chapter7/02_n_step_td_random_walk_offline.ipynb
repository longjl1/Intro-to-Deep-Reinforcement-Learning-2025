# Chapter 7: n-step TD on the Random Walk (Offline)

```python
import numpy as np

def generate_episode(rng):
    s = 3
    states = [s]
    rewards = [0.0]
    while s not in (0, 6):
        s = s + rng.choice([-1, 1])
        states.append(s)
        rewards.append(1.0 if s == 6 else 0.0)
    return states, rewards

rng = np.random.default_rng(1)
n = 4
alpha = 0.1
V = np.full(7, 0.5)
V[0] = V[6] = 0.0
episodes = [generate_episode(rng) for _ in range(50)]
for states, rewards in episodes:
    T = len(states) - 1
    updates = np.zeros_like(V)
    counts = np.zeros_like(V)
    for tau in range(T):
        G = sum(rewards[i] for i in range(tau + 1, min(tau + n, T) + 1))
        if tau + n < T:
            G += V[states[tau + n]]
        updates[states[tau]] += G
        counts[states[tau]] += 1
    for s in range(1, 6):
        if counts[s] > 0:
            target = updates[s] / counts[s]
            V[s] += alpha * (target - V[s])
V
```
