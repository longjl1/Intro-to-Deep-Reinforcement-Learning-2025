# Chapter 10: R-learning on the Access-Control Queuing Task

```python
import numpy as np

Q = np.zeros((11, 4, 2))
rho = 0.0
alpha = 0.1
beta = 0.01

free = 6
priority_idx = 2
action = 1
reward = 4
next_free = 5

target = reward - rho + Q[next_free].max()
Q[free, priority_idx, action] += alpha * (target - Q[free, priority_idx, action])
rho += beta * (reward - rho + Q[next_free].max() - Q[free, priority_idx].max())
Q[free, priority_idx], rho
```

R-learning is framed around average reward rather than discounted return, which makes it a natural fit for continuing tasks such as access-control queuing.
