# Chapter 7: n-step TD on the Random Walk (Online)

```python
import numpy as np

def generate_episode(rng):
    s = 3
    states = [s]
    rewards = [0.0]
    while s not in (0, 6):
        s = s + rng.choice([-1, 1])
        states.append(s)
        rewards.append(1.0 if s == 6 else 0.0)
    return states, rewards

rng = np.random.default_rng(0)
n = 3
alpha = 0.1
V = np.full(7, 0.5)
V[0] = V[6] = 0.0

for _ in range(100):
    states, rewards = generate_episode(rng)
    T = len(states) - 1
    for t in range(T):
        tau = t - n + 1
        if tau >= 0:
            G = sum(rewards[i] for i in range(tau + 1, min(tau + n, T) + 1))
            if tau + n < T:
                G += V[states[tau + n]]
            V[states[tau]] += alpha * (G - V[states[tau]])
V
```
