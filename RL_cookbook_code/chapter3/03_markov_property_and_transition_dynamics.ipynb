## Theory

The Markov property says that once we know the current state and action, the next-state and reward distribution is fully determined probabilistically.

That distribution is written as:

$$
p(s', r \mid s, a).
$$

For each fixed $(s, a)$, the probabilities over all possible next outcomes must sum to one.

This compact object contains the full one-step dynamics of the environment.

```python
import numpy as np
```

```python
transition_model = {
    ('A', 'stay'): [('A', -1.0, 1.0)],
    ('A', 'move'): [('B', 0.0, 0.8), ('A', -1.0, 0.2)],
    ('B', 'stay'): [('B', -0.5, 1.0)],
    ('B', 'move'): [('C', 2.0, 0.7), ('B', -0.5, 0.3)],
}

for key, outcomes in transition_model.items():
    total_prob = sum(prob for _, _, prob in outcomes)
    print(f'{key} total probability = {total_prob:.1f}')
```

```python
def expected_immediate_reward(state, action):
    return sum(reward * prob for _, reward, prob in transition_model[(state, action)])

for pair in transition_model:
    print(pair, expected_immediate_reward(*pair))
```

## Takeaway

- The Markov property is what makes state-based reasoning possible.
- The transition model is a compact description of one-step uncertainty.
- Bellman equations work because they build directly on these one-step dynamics.
