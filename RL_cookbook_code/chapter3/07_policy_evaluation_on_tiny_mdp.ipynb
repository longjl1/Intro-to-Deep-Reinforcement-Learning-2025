# Sutton & Barto Chapter 3: Policy Evaluation on a Tiny MDP
## Estimating Values Under a Fixed Policy

This notebook gives a small, explicit example of policy evaluation in a finite MDP. The focus is still conceptual: we already know the model and the policy, and we want to estimate how good each state is.

## What You Will See

- a two-state non-terminal MDP
- repeated Bellman expectation updates
- convergence of state values under a fixed policy

Dependencies: `numpy`, `matplotlib`

## Theory

Policy evaluation means computing or approximating $v_\pi(s)$ for all states under a fixed policy $\pi$.

In a small finite MDP with known dynamics, we can repeatedly apply the Bellman expectation update until the values stabilize. This is a simple preview of the dynamic programming methods that appear more explicitly in later chapters.

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')
```

```python
states = ['A', 'B', 'T']
gamma = 0.9

policy = {
    'A': {'left': 0.5, 'right': 0.5},
    'B': {'left': 0.3, 'right': 0.7},
}

transitions = {
    ('A', 'left'): ('A', -1.0),
    ('A', 'right'): ('B', 0.0),
    ('B', 'left'): ('A', 0.0),
    ('B', 'right'): ('T', 2.0),
}

def evaluate(iterations=30):
    v = {s: 0.0 for s in states}
    history = []
    for _ in range(iterations):
        new_v = v.copy()
        for s in ['A', 'B']:
            total = 0.0
            for a, prob in policy[s].items():
                next_state, reward = transitions[(s, a)]
                total += prob * (reward + gamma * v[next_state])
            new_v[s] = total
        v = new_v
        history.append([v['A'], v['B']])
    return v, np.array(history)

v_pi, history = evaluate()
v_pi
```

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.plot(history[:, 0], label='v(A)')
ax.plot(history[:, 1], label='v(B)')
ax.set_title('Policy Evaluation by Repeated Bellman Updates')
ax.set_xlabel('Iteration')
ax.set_ylabel('Value')
ax.legend()
plt.tight_layout()
```

## Takeaway

- Policy evaluation studies one policy at a time.
- Repeated Bellman expectation updates can converge to the state values of that policy.
- Later chapters turn this idea into systematic algorithms.
