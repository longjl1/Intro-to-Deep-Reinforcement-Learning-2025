# Chapter 3 Formula Sheet
## Finite Markov Decision Processes

This notebook is a compact formula summary for the Chapter 3 concepts used throughout the rest of the folder.

## Return

$$
G_t = R_{t+1} + \gamma R_{t+2} + \gamma^2 R_{t+3} + \cdots
$$

## State-value function

$$
v_\pi(s) = \mathbb{E}_\pi[G_t \mid S_t = s]
$$

## Action-value function

$$
q_\pi(s, a) = \mathbb{E}_\pi[G_t \mid S_t = s, A_t = a]
$$

## Bellman expectation equation

$$
v_\pi(s) = \sum_a \pi(a \mid s) \sum_{s', r} p(s', r \mid s, a)[r + \gamma v_\pi(s')]
$$

## Action-value form

$$
q_\pi(s, a) = \sum_{s', r} p(s', r \mid s, a)[r + \gamma \sum_{a'} \pi(a' \mid s') q_\pi(s', a')]
$$

## Bellman optimality equation

$$
v_*(s) = \max_a \sum_{s', r} p(s', r \mid s, a)[r + \gamma v_*(s')]
$$

$$
q_*(s, a) = \sum_{s', r} p(s', r \mid s, a)[r + \gamma \max_{a'} q_*(s', a')]
$$


## Takeaway

These equations define the conceptual backbone of finite MDPs and are the main mathematical objects Chapter 3 introduces before the algorithmic chapters begin.
