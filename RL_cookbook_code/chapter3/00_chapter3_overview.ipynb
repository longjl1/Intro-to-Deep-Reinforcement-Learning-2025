# Sutton & Barto Chapter 3 Overview
## Finite Markov Decision Processes at a Glance

This notebook is a short roadmap for the Chapter 3 materials in this folder.

Chapter 3 is mostly conceptual. Instead of presenting a large algorithm, it builds the language that later reinforcement learning methods rely on: states, actions, transitions, returns, value functions, and Bellman equations.

## Core Ideas

A finite MDP formalizes sequential decision making with:

- a finite set of states
- a finite set of actions
- transition probabilities
- rewards
- a discount factor

Once that model is defined, we can ask:

- how good is a state under a policy?
- how good is a particular action in a state?
- what does the best possible policy look like?
- how can one-step lookahead improve a policy?
