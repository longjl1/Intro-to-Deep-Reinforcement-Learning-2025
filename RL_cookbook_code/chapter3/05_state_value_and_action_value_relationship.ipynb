# Sutton & Barto Chapter 3: State Values and Action Values
## Relating $v_\pi(s)$ and $q_\pi(s, a)$

This notebook isolates one very useful Chapter 3 idea: the relationship between state-value functions and action-value functions under a fixed policy.

## What You Will See

- how $v_\pi$ averages over actions under a policy
- how $q_\pi$ performs one-step lookahead from a state-action pair
- a tiny numerical example that computes both objects

Dependencies: `numpy`

## Theory

For a fixed policy $\pi$:

$$
v_\pi(s) = \sum_a \pi(a \mid s) q_\pi(s, a).
$$

This says the value of a state is the policy-weighted average of the values of its available actions.

In the other direction:

$$
q_\pi(s, a) = \sum_{s', r} p(s', r \mid s, a)[r + \gamma v_\pi(s')].
$$

So $q_\pi$ can be computed from one-step dynamics and the next-state values.

```python
gamma = 0.9

policy = {
    'A': {'left': 0.25, 'right': 0.75},
    'B': {'left': 0.60, 'right': 0.40},
}

transitions = {
    ('A', 'left'): ('A', -1.0),
    ('A', 'right'): ('B', 0.0),
    ('B', 'left'): ('A', 0.0),
    ('B', 'right'): ('T', 2.0),
}

v_pi = {'A': 1.2, 'B': 1.8, 'T': 0.0}
```

```python
q_pi = {}
for (s, a), (next_state, reward) in transitions.items():
    q_pi[(s, a)] = reward + gamma * v_pi[next_state]

q_pi
```

```python
v_from_q = {}
for s in policy:
    v_from_q[s] = sum(policy[s][a] * q_pi[(s, a)] for a in policy[s])

v_from_q
```

## Takeaway

- $q_\pi$ looks one step ahead and then uses state values.
- $v_\pi$ averages action values according to the policy.
- This two-way relationship is a convenient way to move between policy evaluation and action selection.
