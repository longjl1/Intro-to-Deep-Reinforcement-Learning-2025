```python
import numpy as np

alpha = 0.3
gamma = 1.0
lam = 0.9
n_features = 64
weights = np.zeros(n_features)
eligibility = np.zeros(n_features)

def active_features(position, velocity, action):
    base = int(((position + 1.2) / 1.7) * 7) + 8 * int(((velocity + 0.07) / 0.14) * 7)
    return [(base + 5 * action + i) % n_features for i in range(4)]

phi = active_features(-0.5, 0.01, 2)
phi
```

Eligibility traces let recent features share credit across multiple time steps, which is why the method can learn faster than one-step Sarsa in many settings.
