# Chapter 10: n-step Sarsa on Mountain Car

```python
import numpy as np

n = 4
alpha = 0.2
weights = np.zeros(128)

def q_hat(position, velocity, action):
    idx = int((position + 1.2) / 1.7 * 15) + 16 * int((velocity + 0.07) / 0.14 * 7)
    features = [(idx + action * 7 + i) % len(weights) for i in range(8)]
    return weights[features].sum(), features

q_hat(-0.5, 0.01, 1)
```

Larger n can propagate reward information farther before bootstrapping, but it also increases the amount of waiting and variance in the update target.
