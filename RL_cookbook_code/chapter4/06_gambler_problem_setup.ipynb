# Chapter 4: Gambler's Problem Setup

The gambler's problem is a classic episodic dynamic programming example from Sutton and Barto.

A state is the gambler's capital. Each action is a stake. If the coin comes up heads, capital increases by the stake; otherwise it decreases by the stake.

The goal is to reach the terminal capital of 100.

```python
goal = 100
p_head = 0.4

def actions(state):
    return list(range(1, min(state, goal - state) + 1))

actions(50)[:10], len(actions(50))
```
