# Sutton & Barto Chapter 4 Overview
## Dynamic Programming for Finite MDPs

This notebook is a roadmap for the Chapter 4 materials in this folder.

Chapter 4 turns the value-function ideas from Chapter 3 into concrete planning procedures for finite MDPs with known models.

## Core Themes

- policy evaluation
- policy improvement
- policy iteration
- value iteration
- generalized policy iteration
- small planning case studies

Start with policy evaluation and policy improvement, then move to policy iteration and value iteration. The gambler's problem and Jack's car rental are good examples to revisit after the core planning ideas are clear.
