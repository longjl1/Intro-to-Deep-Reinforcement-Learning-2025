# Chapter 4: Gambler's Problem with Value Iteration

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')

goal = 100
p_head = 0.4
v = np.zeros(goal + 1)
v[goal] = 1.0

def possible_actions(s):
    return range(1, min(s, goal - s) + 1)

for _ in range(100):
    delta = 0.0
    for s in range(1, goal):
        old = v[s]
        v[s] = max(p_head * v[s + a] + (1 - p_head) * v[s - a] for a in possible_actions(s))
        delta = max(delta, abs(old - v[s]))
    if delta < 1e-9:
        break

policy = np.zeros(goal + 1, dtype=int)
for s in range(1, goal):
    qs = [p_head * v[s + a] + (1 - p_head) * v[s - a] for a in possible_actions(s)]
    policy[s] = list(possible_actions(s))[int(np.argmax(qs))]

fig, ax = plt.subplots(1, 2, figsize=(12, 4))
ax[0].plot(v)
ax[0].set_title('Optimal value function')
ax[1].plot(policy)
ax[1].set_title('Greedy stake policy')
plt.tight_layout()
```

## Interpretation

The policy is often jagged rather than smooth. That is a classic feature of the gambler's problem and comes from the discrete actions available at each capital level.

If the coin becomes fairer or more favorable, the shape of the optimal policy changes noticeably because aggressive stakes become relatively more attractive.
