# Chapter 4: Policy Evaluation Basics

Dynamic programming starts by computing how good a fixed policy is when the environment model is known.

## Theory

Given a policy $\pi$, policy evaluation applies Bellman expectation updates until the values stabilize:

$$
v_{k+1}(s) = \sum_a \pi(a \mid s) \sum_{s', r} p(s', r \mid s, a)[r + \gamma v_k(s')].
$$

This is often called **iterative policy evaluation**.

```python
gamma = 0.9
states = ['A', 'B', 'T']
policy = {'A': {'left': 0.5, 'right': 0.5}, 'B': {'left': 0.5, 'right': 0.5}}
transitions = {
    ('A', 'left'): ('A', -1.0),
    ('A', 'right'): ('B', 0.0),
    ('B', 'left'): ('A', 0.0),
    ('B', 'right'): ('T', 2.0),
}

v = {s: 0.0 for s in states}
for _ in range(20):
    new_v = v.copy()
    for s in ['A', 'B']:
        total = 0.0
        for a, prob in policy[s].items():
            ns, r = transitions[(s, a)]
            total += prob * (r + gamma * v[ns])
        new_v[s] = total
    v = new_v
v
```
